```python
import  pandas as pd
import numpy as np

df = pd.read_csv("/content/HR DATA.csv")
df.head()
```

```python
grouped_stat = df.groupby("Attrition")[["Age","MonthlyIncome"]].agg(["mean","median","min","max","std"])

print(grouped_stat)
```

```python
income_yes = df[df["Attrition"] == "Yes"]["MonthlyIncome"].tolist()
income_no = df[df["Attrition"] == "No"]["MonthlyIncome"].tolist()

print("Income List for Attrition = Yes:", income_yes[:5])
print("Income List for Attrition = No:", income_no[:5])
```

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
print("Statitistics For Iris-setos")
print(iris[iris["Species"]=="Iris-setosa"].describe())

print("\n Statistics for Iris-versicolor")
print(iris[iris["Species"]=="Iris-versicolor"].describe())

print("\n Statistics for Iris-virginica")
print(iris[iris["Species"]=="Iris-virginica"].describe())
```

```python
for Species in iris["Species"].unique():
  print("\nSpecies : ",Species)
  subset = iris[iris["Species"] == Species]
  print("Mean:\n", subset.mean(numeric_only=True))
  print("Median:\n", subset.median(numeric_only=True))
  print("Standard Deviation:\n", subset.std(numeric_only=True))

```
