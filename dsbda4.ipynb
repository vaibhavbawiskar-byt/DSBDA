```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

data_features = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target_values = raw_df.values[1::2, 2]

feature_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'
]

from sklearn.utils import Bunch
boston = Bunch(
    data=data_features,
    target=target_values,
    feature_names=feature_names,
    DESCR="Boston Housing Dataset (fetched from original source due to removal from scikit-learn)"
)
```

```python
data = pd.DataFrame(boston.data, columns=boston.feature_names)
display(data.head())
```

```python
data['PRICE'] = boston.target
display(data.head())
```

```python
print(data.isnull().sum())
```

```python
x = data.drop(['PRICE'], axis=1)
y = data['PRICE']
print("Shape of X:", x.shape)
print("Shape of Y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=0)
print("Shape of xtrain:", xtrain.shape)
print("Shape of xtest:", xtest.shape)
print("Shape of ytrain:", ytrain.shape)
print("Shape of ytest:", ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain, ytrain)
```

```python
ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)
```

```python
df_train_preds = pd.DataFrame({'Actual': ytrain, 'Predicted': ytrain_pred})
df_test_preds = pd.DataFrame({'Actual': ytest, 'Predicted': ytest_pred})

print("\n--- Training Data Predictions ---")
display(df_train_preds.head())
print("\n--- Test Data Predictions ---")
display(df_test_preds.head())
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse_test = mean_squared_error(ytest, ytest_pred)
r2_test = r2_score(ytest, ytest_pred)

mse_train = mean_squared_error(ytrain, ytrain_pred)
r2_train = r2_score(ytrain, ytrain_pred)

print(f"Mean Squared Error (Test): {mse_test:.4f}")
print(f"R-squared (Test): {r2_test:.4f}")
print(f"Mean Squared Error (Train): {mse_train:.4f}")
print(f"R-squared (Train): {r2_train:.4f}")
```
