```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset = sns.load_dataset('titanic')

sns.histplot(data=dataset, x='fare', kde=False, bins=10)
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=dataset)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.jointplot(x='age', y='fare', data=dataset, kind='scatter')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.swarmplot(x='pclass', y='age', hue='survived', data=dataset)
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
